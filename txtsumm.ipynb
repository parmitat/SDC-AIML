```python
# Install the required library
!pip install transformers

# Import required modules
from transformers import pipeline

# Load the summarization pipeline
summarizer = pipeline("summarization")

# Define your text to be summarized
text = """
Artificial Intelligence (AI) is rapidly evolving and impacting various industries.
It enables machines to learn from data, make decisions, and perform tasks that traditionally required human intelligence.
Applications of AI include natural language processing, computer vision, robotics, and more.
AI has the potential to revolutionize the way we work, live, and interact with technology.
"""

# Summarize the text
summary = summarizer(text, max_length=50, min_length=25, do_sample=False)

# Print the summary
print("Original Text:")
print(text)
print("\nSummary:")
print(summary[0]['summary_text'])
```
